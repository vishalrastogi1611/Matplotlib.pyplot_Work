# Groupby value of each catagory

```python
sba = df.groupby(['Customer Segment']).agg({'Sales':'sum'}).reset_index()
```

```python
sba
```

```python
plt.figure(figsize=(6,2))
plt.bar(sba['Customer Segment'],sba['Sales'])
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.bar(sba['Customer Segment'],sba['Sales'])
plt.show()
```

```python
sba['Sales']=sba['Sales']/1000
```

```python
plt.figure(figsize=(6,2))
plt.bar(sba['Customer Segment'],sba['Sales'])
plt.show()
```

```python
# Find mean
```

```python
ba = df.groupby(['Customer Segment']).agg({'Sales':'mean'}).reset_index()
```

```python
ba
```

```python
plt.figure(figsize=(6,2))
plt.plot(sba['Customer Segment'],sba['Sales'],marker='o',mfc='r',mec='y',ms=10,ls='--',color='red',linewidth=2)
plt.title('Sales by segment')
plt.ylabel('total sales in K')
plt.xlabel('segment cat..')
plt.show()
```
