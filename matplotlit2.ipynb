# Subploting

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Vishal/Documents/Powerbidatas/shopping_trends.csv").ffill()
```

```python
df.head(3)
```

```python
pbc = df.groupby(['Category']).agg({'Purchase Amount (USD)':'sum'}).reset_index()
sbc = df.groupby(['Season']).agg({'Purchase Amount (USD)':'sum'}).reset_index()
cbc = df.groupby(['Payment Method']).agg({'Purchase Amount (USD)':'sum'}).reset_index()
ppc = df.groupby(['Category']).agg({'Previous Purchases':'sum'}).reset_index()
```

```python
pbc,  sbc, cbc , ppc
```

```python
plt.figure(figsize=(4,4))
plt.subplot(2,1,1)
plt.plot(pbc['Category'],pbc['Purchase Amount (USD)'])
plt.subplot(2,1,2)
plt.bar(sbc['Season'],sbc['Purchase Amount (USD)'])
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.subplot(1,2,1)
plt.plot(pbc['Category'],pbc['Purchase Amount (USD)'],color='g')
plt.subplot(1,2,2)
plt.bar(sbc['Season'],sbc['Purchase Amount (USD)'],color='red')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(2,2,1)
plt.plot(pbc['Category'],pbc['Purchase Amount (USD)'],color='g')
plt.subplot(2,2,2)
plt.bar(sbc['Season'],sbc['Purchase Amount (USD)'],color='red')
plt.subplot(2,2,3)
plt.bar(cbc['Payment Method'],cbc['Purchase Amount (USD)'],color='y')
plt.subplot(2,2,4)
plt.plot(ppc['Category'],ppc['Previous Purchases'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.pie(pbc['Purchase Amount (USD)'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.pie(pbc['Purchase Amount (USD)'],labels=pbc['Category'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
exp = (0,0,0,0.2)
plt.pie(pbc['Purchase Amount (USD)'],labels=pbc['Category'],explode=exp)
plt.show()
```

```python
plt.figure(figsize=(4,4))
exp = (0,0,0,0.2)
plt.pie(pbc['Purchase Amount (USD)'],labels=pbc['Category'],explode=exp,shadow=True)
plt.show()
```
